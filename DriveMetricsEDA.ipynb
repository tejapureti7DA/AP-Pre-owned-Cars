```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.countplot(
    y='brand',
    data=df,
    order=df['brand'].value_counts().index
)

plt.title("brand distribution")
plt.xlabel("number of cars")
plt.ylabel("brand")
plt.show()
```

```python
df['fuel_type'].value_counts()
```

```python
fuel_count = df['fuel_type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    fuel_count,
    labels=fuel_count.index,
    autopct='%1.1f%%'
)

plt.title("fuel type distribution")
plt.show()
```

```python
district_revenue = df.groupby(
    'district'
)['selling_price'].sum()
```

```python
plt.figure(figsize=(12,6))

district_revenue.sort_values(
    ascending=False
).plot(kind='bar')

plt.title("revenue by district")
plt.xlabel("district")
plt.ylabel("revenue")
plt.show()
```

```python
df['vehicle_age'] = 2025 - df['year']
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='vehicle_age',
    y='selling_price',
    data=df
)

plt.title("vehicle age vs selling price")
plt.xlabel("vehicle age")
plt.ylabel("selling price")
plt.show()
```

```python
df['sale_date'] = pd.to_datetime(
    df['sale_date']
)
```

```python
df['month'] = df['sale_date'].dt.month_name()
```

```python
monthly_sales = df.groupby(
    'month'
)['car_id'].count()
```

```python
plt.figure(figsize=(10,5))

monthly_sales.plot(
    kind='line',
    marker='o'
)

plt.title("monthly sales trend")
plt.xlabel("month")
plt.ylabel("cars sold")
plt.show()
```

```python
brand_share = df['brand'].value_counts()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    brand_share,
    labels=brand_share.index,
    autopct='%1.1f%%'
)

centre_circle = plt.Circle(
    (0,0),
    0.70,
    fc='white'
)

fig = plt.gcf()
fig.gca().add_artist(
    centre_circle
)

plt.title("brand market share")
plt.show()
```
